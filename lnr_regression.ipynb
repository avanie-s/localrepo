```python
import tensorflow as tf
import matplotlib.pyplot as plt
import numpy as np
import keras
```

```python

#LINEAR REGRESSION
# y = 5x + 6

x_train = tf.constant([0.,2.,-1.,3.,6.])
y_train = tf.constant([6.,16.,1.,21.,36.])

w = tf.Variable(0.0)
b = tf.Variable(0.0)
mse_loss = tf.keras.losses.MeanSquaredError()
lr = tf.constant(0.01)#learning rate

for steps in range(1000):
    with tf.GradientTape() as tape:
        y_pred = w*x_train + b
        loss = mse_loss(y_train, y_pred)   
        dw, db = tape.gradient(loss, [w, b])
        w.assign(w-(lr*dw))
        b.assign(b-(lr*db))

x_test = tf.constant([7.,8.,-4.,1.,5.])
y_test = x_test*w + b
plt.plot(x_test,y_test, color='pink')
plt.title("y = " + str(w.numpy()) + " * x + " + str(b.numpy()))
plt.grid(True)
plt.show()

```
